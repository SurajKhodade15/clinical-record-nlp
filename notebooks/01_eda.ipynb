```python
## Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cl_df = pd.read_csv('../data/clinical_dataset.csv')
```

```python
cl_df.head()
```

```python
cl_df.describe()
```

```python
cl_df[cl_df['Billing Amount'] < 0]
```

```python
# ================================
# 12. INVALID BILLING AMOUNTS CHECK
# ================================

# Identify records with negative billing amounts
invalid_billing = cl_df[cl_df['Billing Amount'] < 0]
print(f"Records with negative billing amount: {invalid_billing.shape[0]}")

# Display sample
display(invalid_billing.head())

# Visualize
plt.figure(figsize=(6, 4))
sns.histplot(cl_df['Billing Amount'], bins=50, kde=True)
plt.axvline(0, color='red', linestyle='--', label='Zero Threshold')
plt.title("Billing Amount Distribution with Negative Values Highlighted")
plt.xlabel("Billing Amount")
plt.legend()
plt.tight_layout()
plt.savefig('../plots/eda/billing_amount_negative_check.png')
plt.show()
```

```python
cl_df['Billing Amount'].mean()
```

```python
median_billing = cl_df[cl_df['Billing Amount'] >= 0]['Billing Amount'].median()
cl_df.loc[cl_df['Billing Amount'] < 0, 'Billing Amount'] = median_billing
```

```python
cl_df[cl_df['Billing Amount'] < 0]
```

```python
missing = cl_df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]

plt.figure(figsize=(10, 6))
sns.barplot(x=missing.values, y=missing.index)
plt.title('Missing Value Count per Column')
plt.xlabel('Missing Values')
plt.ylabel('Columns')
plt.tight_layout()
plt.savefig('../plots/eda/missing_values.png')
plt.show()
```

```python
# Numerical Features
numerical_cols = ['Age', 'Billing Amount']
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(cl_df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.savefig(f'../plots/eda/{col.lower().replace(" ", "_")}_distribution.png')
    plt.show()
```

```python
categorical_cols = ['Gender', 'Blood Type', 'Medical Condition', 
                     
                    'Admission Type', 'Medication', 'Test Results']

for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    cl_df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.savefig(f'../plots/eda/{col.lower().replace(" ", "_")}_distribution.png')
    plt.show()
```

```python
# Convert dates
cl_df['Date of Admission'] = pd.to_datetime(cl_df['Date of Admission'])
cl_df['Discharge Date'] = pd.to_datetime(cl_df['Discharge Date'])

# Admission Trend
admission_trend = cl_df['Date of Admission'].dt.to_period('M').value_counts().sort_index()

plt.figure(figsize=(12, 5))
admission_trend.plot()
plt.title('Monthly Patient Admissions')
plt.ylabel('Number of Admissions')
plt.xlabel('Month')
plt.tight_layout()
plt.savefig('../plots/eda/monthly_admissions_trend.png')
plt.show()

# ================================
# 9. CORRELATION ANALYSIS
# ================================
plt.figure(figsize=(6, 4))
sns.heatmap(cl_df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig('../plots/eda/correlation_heatmap.png')
plt.show()

# ================================
# 10. OUTLIER DETECTION
# ================================
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=cl_df[col])
    plt.title(f'Outliers in {col}')
    plt.tight_layout()
    plt.savefig(f'../plots/eda/{col.lower().replace(" ", "_")}_boxplot.png')
    plt.show()
```

```python
cl_df.to_csv('../data/clinical_dataset_cleaned.csv', index=False)
```
